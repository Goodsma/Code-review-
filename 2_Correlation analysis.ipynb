In feature selection, In general, there are three types of techniuqe:

Filter methods
The simplest approach is to find a single measure that relates each independent variable to the dependent variable, which can be used to score the importance of the independent variable.
Correlation is a particularly common measure to do this, but others exist as well.

Embedded approaches
Some algorithms have built-in feature selection due to the way they model data.
For example, decision trees and lasso regression.


Wrapper approaches
It is also possible to generate a vast number of models with various subsets of independent variables to check what subsets give the best performance.
The benefit of doing this is that, by testing various subsets, various combinations are tested as well. Hence, the interaction between different variables is also taken into account.

Spearman’s Rank correlation coefficient
It also does not just test for linear relationships and is suited for both ordinal and continuous variables.

For discrete observations, another measure of association that is commonly used is the Pearson’s -statistic: 
The value ranges between 0 and the size of the number of observations () times the minimum of the number of values either variable can take () minus 1. This gives you a clue about how strong the relationship is compared to how strong it can be.
The Cramer’s V-statistic normalizes the -statistic accordingly:
The closer this metric is to one, the stronger the relationship.

Finally, there is the mutual information metric
It is also a measure of similarity between two different variables.


# Correlation analysis

Now we will calculate the correlation of the variables in a dataset. First, we import the data into pandas:

## The dataset

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from pandas.plotting import scatter_matrix
from sklearn import datasets

dataset = datasets.load_boston()
var_names = dataset['feature_names']
X = dataset['data']
y = dataset['target']
```

```python
print(dataset['DESCR'])
```

```python
data = pd.DataFrame(X)
data.columns = var_names
data['TARGET'] = pd.DataFrame(y)
print(data.head())
```

## Calculating the correlation

Now, we calculate the correlation of the (continuous) variables:

```python
correlation = data.corr()
print(correlation)
```

Let's visualize a single relationship:

```python
plt.scatter(X[5],X[6])
plt.xlabel(data.columns[5])
plt.ylabel(data.columns[6])
plt.show()
```

Do you notice that there are clearly outliers present?

## The correlation plot

Next, let's visualise this with a correlation plot:

```python
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlation, vmin=-1, vmax=1)
fig.colorbar(cax)

ticks = np.arange(0,14,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# This rotates the labels on the x-axis
for tick in ax.get_xticklabels():
    tick.set_rotation(90)

var_names = np.append(var_names, 'TARGET')

ax.set_xticklabels(var_names)
ax.set_yticklabels(var_names)
plt.show()
```

We will now get an idea of the distribution of the variables(please note,this code might run for a while):

```python
scatter_matrix(data)
plt.show()
```

## Other measures

Finally, let's compute 𝜒2 and mutual information measures. They are typically used in conjunction with the SelectKBest function:

```python
from sklearn.feature_selection import chi2, mutual_info_regression, mutual_info_classif
from sklearn.feature_selection import SelectKBest

print('Mutual information:')

selected_features = SelectKBest(mutual_info_regression, k=4).fit(data, y)
for feature_name, selected in zip(data.columns, selected_features.get_support()):
    print(feature_name+' is selected '+str(selected))
```

Given that y is continuous, we first bin it for illustration purposes

```python
y_binned = pd.cut(data['TARGET'],3,labels=['bin1','bin2','bin3'])
```

```python
print('Chi-squared:')
selected_features = SelectKBest(chi2, k=4).fit(data, y_binned)
for feature_name, selected in zip(data.columns, selected_features.get_support()):
    print(feature_name+' is selected '+str(selected))
```

```python
print('Mutual information:')
selected_features = SelectKBest(mutual_info_classif, k=4).fit(data, y_binned)
for feature_name, selected in zip(data.columns, selected_features.get_support()):
    print(feature_name+' is selected '+str(selected))
```
